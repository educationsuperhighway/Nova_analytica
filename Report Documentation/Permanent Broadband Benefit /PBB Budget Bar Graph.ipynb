```python

# Plotting
fig, ax1 = plt.subplots(figsize=(12,6))

# Stacked bar chart
bar_width = 0.6
bars1 = plt.bar(years, potential_reallocation, color='#009296',alpha=0.4, width=bar_width, label='Potential for Reallocation')
bars2 = plt.bar(years, committed_funds, bottom=potential_reallocation, color='#4a4a4a',alpha=0.2, width=bar_width, label='Committed to High Cost Fund')

# Line charts
# Line charts, using np.nan for None values
pbb_scenario_1 = [np.nan if v is None else v for v in eligible_26_3M]
pbb_scenario_2 = [np.nan if v is None else v for v in eligible_19_3M]
pbb_scenario_3 = [np.nan if v is None else v for v in eligible_16_3M]

ax1.plot(years, pbb_scenario_1, color='#F26B21', label='26.3M Eligible')
ax1.plot(years, pbb_scenario_2, color='#F09221', label='19.3M Eligible')
ax1.plot(years, pbb_scenario_3, color='#FDB913', label='16.3M Eligible')
# Add data labels
for i, txt in enumerate(pbb_scenario_1):
    if not np.isnan(txt):
        label = f"${txt/1e9:.1f}B" if txt >= 1e9 else f"${txt/1e6:.0f}M"
        ax1.annotate(label, (years[i], pbb_scenario_1[i]), textcoords="offset points", xytext=(0,10), ha='center', 
                     fontsize=9, weight='bold', fontname = 'Arial', color = 'black')

for i, txt in enumerate(pbb_scenario_2):
    if not np.isnan(txt):
        label = f"${txt/1e9:.1f}B" if txt >= 1e9 else f"${txt/1e6:.0f}M"
        ax1.annotate(label, (years[i], pbb_scenario_2[i]), textcoords="offset points", xytext=(0,10), ha='center', 
                     fontsize=9, weight='bold', fontname = 'Arial', color = 'black')

for i, txt in enumerate(pbb_scenario_3):
    if not np.isnan(txt):
        label = f"${txt/1e9:.1f}B" if txt >= 1e9 else f"${txt/1e6:.0f}M"
        ax1.annotate(label, (years[i], pbb_scenario_3[i]), textcoords="offset points", xytext=(0,-10), ha='center', 
                     fontsize=9, weight='bold', fontname = 'Arial', color = 'black')

# Labels and title
plt.xlabel('Year')
plt.ylabel('Billions of Dollars')
plt.suptitle('High Cost Commitments vs. Projected PBB Budgets', size=16)
plt.title('Three PBB Budget scenarios reflect different eligibility assumptions', size=12)
#label years, but remove tick marks.
plt.xticks(years)
ax1.tick_params(axis='both', which='both', bottom=False)

plt.legend(loc='upper left', prop={'size': 10})
#remove frame
plt.box(False)

#add anntoations
ax1.annotate('RDOF Funds Expire',
             xy=(2031, potential_reallocation[years.index(2031)]),
             xytext=(-20,18),  # Offset the text above the point
             textcoords='offset points',
             ha='center', fontsize=10, color='#009296', fontweight='bold',
            arrowprops=dict(facecolor='#009296', width = 4, headwidth=10, headlength=4))

ax1.annotate('E-ACAM Funds Expire',
             xy=(2039, potential_reallocation[years.index(2039)]),
             xytext=(-20,18),  # Offset the text above the point
             textcoords='offset points',
             ha='center', fontsize=10, color='#009296', fontweight='bold',
            arrowprops=dict(facecolor='#009296', width = 4, headwidth=10, headlength=4))
# Display the plot
plt.show()
```
