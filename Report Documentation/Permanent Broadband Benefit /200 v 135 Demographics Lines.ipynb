```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
d = {'pct_FPL': [200, 135],
    'Overall':[0, -20.13],
    'Seniors':[0, -23.12],
    'Veterans':[0, -25.07],
    'Rural':[0, -22.62],
    'Black':[0, -14],
    'Latinx':[0, -18.16]}

df = pd.DataFrame(data = d)

df
```

```python
#First pass just to see what we're working with

x = df['pct_FPL']
for col in df.columns[1:]:
    plt.plot(x, df[col], label= col)

plt.gca().invert_xaxis()
plt.legend()
```

```python
##Nice looking graph
fig, ax1 = plt.subplots(figsize=(12,6))
#Black
ax1.plot(x, df['Black'], label= 'Black', color='#5F0F40', #purple
        alpha=0.5)
ax1.annotate('14% decrease',
           xy=(135, -14),
             xytext=(38,0),  # Offset the text above the point
             textcoords='offset points',
             ha='center', fontsize=10, color='#5F0F40')

#Latinx
ax1.plot(x, df['Latinx'], label= 'Latinx', color='#F26B21', #tan
        alpha=0.5)
ax1.annotate('18% decrease',
           xy=(135, -18.16),
             xytext=(38,0),  # Offset the text above the point
             textcoords='offset points',
             ha='center', fontsize=10, color='#F26B21')

#Overall
ax1.plot(x, df['Overall'], label= 'All PBB Eligible', color='#4A4A4A', #grey
        linestyle = 'dashed', alpha=0.5)
ax1.annotate('20% decrease',
           xy=(135, -20.13),
             xytext=(38,0),  # Offset the text above the point
             textcoords='offset points',
             ha='center', fontsize=10, color='#4A4A4A')

#Rural
ax1.plot(x, df['Rural'], label= 'Rural', color='#F09221', #orange
        )
ax1.annotate('23% decrease',
           xy=(135, -22.62),
             xytext=(38,2),  # Offset the text above the point
             textcoords='offset points',
             ha='center', fontsize=10, color='#F09221')

#Seniors
ax1.plot(x, df['Seniors'], label= 'Seniors', color='#FDB913', #yellow
        )
ax1.annotate('23% decrease',
           xy=(135, -23.12),
             xytext=(38,-2),  # Offset the text above the point
             textcoords='offset points',
             ha='center', fontsize=10, color='#FDB913')

#Veterans
ax1.plot(x, df['Veterans'], label= 'Veterans', color='#009296', #teal
        )
ax1.annotate('25% decrease',
           xy=(135, -25.07),
             xytext=(38,0),  # Offset the text above the point
             textcoords='offset points',
             ha='center', fontsize=10, color='#009296')

##Formatting the graph
plt.box(False)
plt.legend()
plt.suptitle('Impact of Reducing Eligibility to 135%', size = 16)
plt.title('Percentages taken out of PBB Eligible/Unconnected households at 200% FPL', size = 12)

#x-axis
#invert it
plt.gca().invert_xaxis()
#axis label and ticks
plt.xlabel('Percent of Federal Poverty Line')
ticks_x = list(range(200, 130, -5))
plt.xticks(ticks_x, [f'{i}%' for i in ticks_x]) 

#axis label and ticks
plt.ylabel('Decrease in PBB Eligible/Unconnected Households (%)')
ticks_y = list(range(0, -25, -5))
plt.yticks(ticks_y, [f'{i}%' for i in ticks_y])
```
